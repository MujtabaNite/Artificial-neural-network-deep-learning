```python
import pandas as pd
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt 

```

```python
# load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print(X_train.shape)
print(y_train.shape)
```

```python
X_train = X_train.reshape((60000, 28, 28, 1))
X_train = X_train.astype('float32') / 255.0
X_test = X_test.reshape((10000, 28, 28, 1))
X_test = X_test.astype('float32') / 255.0
```

```python
n_classes = 10
y_train = to_categorical(y_train, n_classes)
y_test = to_categorical(y_test, n_classes)
```

# making and saving the models


```python
cnn = Sequential()
cnn.add(Conv2D(32, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu', input_shape=(28,28,1)))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Conv2D(64, kernel_size=(3,3), activation='relu', padding='same'))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Flatten())
cnn.add(Dense(128, activation='relu'))
cnn.add(Dense(10, activation='softmax'))

# compile with categorical_crossentropy because labels are one-hot
cnn.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9), 
            loss='categorical_crossentropy', 
            metrics=['accuracy'])

cnn.summary()
cnn.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=128, verbose=1)

cnn.save('mnist_CNN.keras')
```

# testing the models

```python
import tensorflow as tf
import cv2
import numpy as np
```

# loading the model

```python
model = tf.keras.models.load_model('mnist_CNN.keras')
```

# resizing the image

```python
img = cv2.imread('digit.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (28, 28))
img = img.reshape(28, 28, 1)
img = img.astype('float32') / 255
img = np.expand_dims(img, axis=0)  # Add batch dimension
```

# predicting using the model

```python
prediction = model.predict(img)
print("Predicted digit:", np.argmax(prediction))
```
