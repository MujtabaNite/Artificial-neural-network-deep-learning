# **Image reading and converting into csv**

```python
import os
import cv2
import csv
```

# ** (28/28/3) **

```python
# Path to dataset 
dataset_path = r"M:\University\University Work\Semestor 5\Ann & DL\LAB\ANN & DL\Data\Vechicle"

# categories
categories = ["airplane", "car", "motorbike"]

# Map labels to numbers (0,1,2,)
label_dict = { "airplane": 0, "car": 1, "motorbike": 2}

# Resize size
IMG_SIZE = 28  # you can change to 64 or 128 (larger = much bigger CSV!)

# Output CSV file
with open("Images28.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write header (label + pixel values)
    header = ["label"] + [f"p{i}" for i in range(IMG_SIZE*IMG_SIZE*3)]
    writer.writerow(header)

    # Process each category
    for emotion in categories:
        path = os.path.join(dataset_path, emotion)
        label = label_dict[emotion]

        for root, dirs, files in os.walk(path):
            for filename in files:
                if filename.lower().endswith(".jpg"):
                    file_path = os.path.join(root, filename)

                    # Read image in COLOR
                    image = cv2.imread(file_path, cv2.IMREAD_COLOR)
                    if image is None:
                        continue
                    
                    # image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Convert to grayscale
                    # Resize to fixed size
                    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))

                    # Flatten (3 channels preserved in sequence R,G,B,...)
                    flat = image.flatten()

                    # Write row: [label, pixel1, pixel2, ...]
                    row = [label] + flat.tolist()
                    writer.writerow(row)

print(" CSV file 'Images28.csv' created successfully.")
```

# 32/32/1 grayscale


```python
# Path to dataset 
dataset_path = r"M:\University\University Work\Semestor 5\Ann & DL\LAB\ANN & DL\Vechicle"

# categories
categories = ["airplane", "car", "motorbike"]

# Map labels to numbers (0,1,2,)
label_dict = { "airplane": 0, "car": 1, "motorbike": 2}

# Resize size
IMG_SIZE = 32  # you can change to 64 or 128 (larger = much bigger CSV!)

# Output CSV file
with open("Images32.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write header (label + pixel values)
    header = ["label"] + [f"p{i}" for i in range(IMG_SIZE*IMG_SIZE*1)]
    writer.writerow(header)

    # Process each category
    for emotion in categories:
        path = os.path.join(dataset_path, emotion)
        label = label_dict[emotion]

        for root, dirs, files in os.walk(path):
            for filename in files:
                if filename.lower().endswith(".jpg"):
                    file_path = os.path.join(root, filename)

                    # Read image in COLOR
                    image = cv2.imread(file_path, cv2.IMREAD_COLOR)
                    if image is None:
                        continue
                    
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Convert to grayscale
                    # Resize to fixed size
                    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))

                    # Flatten (3 channels preserved in sequence R,G,B,...)
                    flat = image.flatten()

                    # Write row: [label, pixel1, pixel2, ...]
                    row = [label] + flat.tolist()
                    writer.writerow(row)

print(" CSV file 'Images32.csv' created successfully.")
```
