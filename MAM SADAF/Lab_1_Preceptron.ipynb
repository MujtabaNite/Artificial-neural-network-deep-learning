```python
import numpy as np

# Preceptron Algorithm variables 

X = np.array([[1,1],[1,0],[0,1],[0,0]])
T = np.array([-1,1,1,-1])
W = np.array([0,0])
b = 0
theta = 0.2
alpha = 1

```

```python
# Perceptron Algorithm
epoch = 0
while epoch < 15:
    print("Epochs:", epoch)
    y_out = []
    for i in range(len(X)):
        v = np.sum(np.dot(W, X[i])) + b

        if v >= theta:
            y = 1
        elif v <= -theta:
            y = -1
        else:
            y = 0

        y_out.append(y)

        if y != T[i]:
            W[0] = W[0] + alpha * T[i] * X[i][0]
            W[1] = W[1] + alpha * T[i] * X[i][1]
            b = b + alpha * T[i]


    epoch += 1
    if np.array_equal(np.array(y_out), T):
        break

print("Final Weights:", W)
print("Final Bias:", b)
print("Final Threshold:", theta)
print("Final Learning Rate:", alpha)
print("Total Epochs:", epoch)
print("Final Outputs:", y_out)
print("Final Targets:", T)
```
