# **Activation funciton**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.arange(-4,4,0.05)
len(z)
z
```

```python
# Sigmoid Function
def sigmoid_Activation(z):
    return 1/(1+np.exp(-z)) , 1/(1+np.exp(-z))*(1-1/(1+np.exp(-z)))


output , derivative = sigmoid_Activation(z)

plt.plot(z,output, label='Sigmoid Activation Function',color='orange')
plt.plot(z,derivative, label='Sigmoid Derivative Function', linestyle='dashed')
plt.title('Sigmoid Activation Function and its Derivative')
plt.xlabel('z')
plt.ylabel('Activation/Derivative')
plt.grid()
plt.show()
plt.savefig('Sigmoid_Activation_Function.png')
```

```python
z1 = np.arange(-4,4,0.05)
```

```python
# Tanh Function
def Tanh_Activation(z):
    s = (2/(1+np.exp(-2*(z)))) - 1
    d = 1 - s**2
    return s,d  


output , derivative = Tanh_Activation(z1)

plt.plot(z,output, label='Tanh Activation Function',color='orange')
plt.plot(z,derivative, label='Tanh Derivative Function', linestyle='dashed')
plt.title('Tanh Activation Function and its Derivative')
plt.xlabel('z')
plt.ylabel('Activation/Derivative')
plt.grid()
plt.show()
plt.savefig('Tanh_Activation_Function.png')
```

```python
# ReLU Function

def ReLU_Activation(z):
    s = np.maximum(0,z)
    d = np.where(z>0,1,0)
    return s,d


output , derivative = ReLU_Activation(z)

plt.plot(z,output, label='ReLU Activation Function',color='orange')
plt.plot(z,derivative, label='ReLU Derivative Function', linestyle='dashed')
plt.title('ReLU Activation Function and its Derivative')
plt.xlabel('z')
plt.ylabel('Activation/Derivative')
plt.grid()
plt.show()
plt.savefig('ReLU.png')
```

```python
z2 = np.array([4,2.5,0.3,1])
# Softmax Function
def Softmax_Activation(z):
    exp  = [np.exp(i) for i in z]
    s = sum(exp)
    sm = [j/s for j in exp]
    return sm

output = Softmax_Activation(z2)
print("Softmax Output: ",output)
```
