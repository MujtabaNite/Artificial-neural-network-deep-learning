```python
 # using opencv

import cv2
m = cv2.imread("img.jpg")
cv2.imshow('mountains',m)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
def display_image(text,i):
    cv2.imshow(text,i)
    cv2.waitkey(0)



img1 = cv2.imread("img.jpg")
display_image("mountains",img1)
```

```python
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
display_image('RGB image',img2)
```

```python
img3 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
display_image('RGB image',img3)
```
