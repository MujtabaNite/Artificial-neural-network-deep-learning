```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np  
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
```

* Convert the data into csv file read thruough pandas and convert its data to numerical
* divide the data to 2 parts 80% training and 20% testing
* apply a suitable classification model
* measure its perfromance accuracy
* plot the salary on x axis and gender on y axis and show a scatter graph with diff color for each class

```python
# 1 
 
data  = pd.read_csv("Dataset.csv")
data.head()
```

```python
le = LabelEncoder()

data["Gender"] = le.fit_transform(data["Gender"] )
data["Peformance"] = le.fit_transform(data["Peformance"])

data.head()
```

```python
X = data.drop(columns=["Peformance"])
y = data["Peformance"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python

model =  DecisionTreeClassifier()
model.fit(X_train, y_train)

predictions = model.predict(X_test)


```

```python
# accuracy 
acc = accuracy_score(y_test, predictions)
acc
```

```python

plt.scatter(data["Salary"], data["Gender"], c=data["Peformance"], cmap='viridis')
plt.xlabel("Salary")
plt.ylabel("Gender")
plt.title("Scatter Plot by Peformance Class")

plt.show()
```
